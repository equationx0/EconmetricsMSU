```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
data = pd.read_csv('ef.csv')
model_1 = smf.ols('Y ~ D', data=data).fit()
model_1.summary()
```

```python
model_2 = smf.ols('Y ~ D + EF', data=data).fit()
model_2.summary() 
```

```python
model_3 = smf.ols('Y ~ D + EF + D * EF', data=data).fit()
model_3.summary()
```
